````python
import pandas as pd
````

````python
import numpy as np
````

````python
df = pd.read_csv('./kc_house_data.csv',encoding='cp1251')
````

````python
df.head(5)
````

### 1.1 Создать новый признак price_per_sq_lot, который будет содержать стоимость за один кв. метр общей площади

````python
df['price_per_sq_lot'] = df['price']/ df['sqft_lot']
````

### 1.2 Создать новый признак delta_renovated, который будет содержать разницу в годах между годом реновацией дома и годом постройки дома. Если реновации дома не было, то в новом признаке поставьте 0

````python
df['delta_renovated'] = df['yr_renovated'] - df['yr_built'] # Создали новуй столбец delta_renovated из разницы годом реновацией дома и годом постройки дома
````

````python

df.loc[df['delta_renovated'] <0, 'delta_renovated'] = 0 # Заменили отрицательные значения на 0
````

### 1.3 Создайте признаки года продажи, месяца продажи

````python
df.head(2)
````

````python
def year_of_sale(stroka):
    return stroka [:4] #  Функция, которая удаляет последние 11 знаков
````

````python
year_of_sale('20141013T000000') #  Проверяю работоспособность функции и о чудо она работает
````

````python
df['Year_Of_Sale'] = df["date"].apply(year_of_sale) # Создали столбец Year_Of_Sale и в нём вставили столбец date к которому применили функцию year_of_sale которая выдала только первые 4 символа
````

````python
df.head(2)
````

````python
def month_of_sale(stroka):
    return stroka [4:][: - 9] # Функция удалаяет первы 4 символа в строке и последние 9 символов в строке
````

````python
month_of_sale('20150218T000000') # Функция работае корректно
````

````python
df['Month_Of_Sale'] = df["date"].apply(month_of_sale) # Создали столбец 'Month_Of_Sale' и в нём вставили столбец date к которому применили функцию month_of_sale которая удалаяет первы 4 символа в строке и последние 9 символов в строке
````

````python
df.head(2)
````
